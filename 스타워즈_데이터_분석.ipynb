## 오류 1

- DF은 split(), strip() 등을 사용할 수 없다. 일단 txt파일 상태에서 '""' 큰 따음표를 기준으로 나눈 후, 각 값을 차례로 리스트에 넣어야 할 것같다.
  - 컬럼은 일단 삭제하고 나중에 DF에 추가하든 한다.
- 각 줄이 리스트로 반환되었다.
`[['"1" "THREEPIO" "Did you hear that?  They\'ve shut down the main reactor.  We\'ll be destroyed for sure.  This is madness!"'][....]`
  - 줄임표인 '가 \를 항상 앞에 달고 있는 것을 볼 수 있는데 이를 해결하기 위해 정규표현식을 사용할 예정이다. `.replace('/\\', '')`
  - 이 리스트들 내의 요소들을 어떻게 나누어야할까?

```python
contents = []
with open('/content/SW_EpisodeIV.txt') as file:
  for line in file: # 줄 단위로 파일을 읽어오기
    line = line.strip().split('""')
    contents.append(line)
del contents[0] # 컬럼제거(character, dialogue)
print(contents)
```

```python
splited_contents = []
for i in contents :
  i.split(',')
  splited_contents.append(i)
print(splited_contents)
```

```python
from io import StringIO
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser

def read_pdf_PDFMINER(pdf_file_path):
    """
    pdf_file_path: 'dir/aaa.pdf'로 구성된 path로부터 
    내부의 text 파일을 모두 읽어서 스트링을 리턴함.
    https://pdfminersix.readthedocs.io/en/latest/tutorials/composable.html
    """
    output_string = StringIO()
    with open(pdf_file_path, 'rb') as f:
        parser = PDFParser(f)
        doc = PDFDocument(parser)
        rsrcmgr = PDFResourceManager()
        device = TextConverter(rsrcmgr, output_string, laparams=LAParams())
        interpreter = PDFPageInterpreter(rsrcmgr, device)
        for page in PDFPage.create_pages(doc):
            interpreter.process_page(page)
    return str(output_string.getvalue())


print(read_pdf_PDFMINER("script-toystory3.pdf"))
# print(type(read_pdf_PDFMINER("script-toystory3.pdf")))
```
